```python
## visualize some random linear data 

import numpy as np
import matplotlib.pyplot as plt

X = 3 * np.random.rand(150,1)
y = 2 + 5*X + np.random.randn(150,1)

plt.figure(figsize=(10,10))
plt.scatter(X,y)
plt.xlabel('X data')
plt.ylabel('Y data')
plt.show()
```

```python
## model as such for optimization
```

```python
## cost function to optimize

def cost_function(X, y, theta):
    n = len(y)
    
    prediction = X.dot(theta)
    cost = (1/2*n) *(sum(prediction-y)**2)
    
    return cost
```

```python
## with a random theta, our cost for the above relationship is quite high
cost_function(X, y, 4)
```

```python
## we can optimize theta via gradient descent

def gradient_descent(X, y, theta, learning_rate = 0.1, iterations = 100000):
    n = len(y)
    cost_hist = []
    theta_hist = []
    
    for i in range(iterations):
        if(i%5000 == 0):
            print(i)
        prediction = X.dot(theta)
        
        ## update weights via learning rate and partial derivative
        theta = theta - (1/n)*learning_rate*(X.T.dot((prediction - y)))
        
        theta_hist.append(float(theta))
        cost_hist.append(cost_function(X, y, theta))
    return theta_hist, cost_hist, theta
```

```python
weights, loss, theta = gradient_descent(X, y,1)
```

```python
plt.figure(figsize = (10,10))
plt.plot(weights, loss)
plt.xlabel('Weight')
plt.ylabel('Loss')
```
