```python
import numpy as np
from skimage import io, viewer
import matplotlib.pyplot as plt
img = io.imread('../../Downloads/download.jpeg', as_grey=True)  # load the image as grayscale
plt.figure(figsize = (9,9))
plt.imshow(img, cmap = plt.cm.gray)
```

```python
## use a sharpening kernel to sharpen the image/accentuate edges

def convolve2d(image, kernel):
    kernel = np.flipud(np.fliplr(kernel))    
    output = np.zeros_like(image)            # convolution output
    ##padding
    image_padded = np.zeros((image.shape[0] + 2, image.shape[1] + 2))   
    image_padded[1:-1, 1:-1] = image
    
    for x in range(image.shape[1]):
        for y in range(image.shape[0]):
            output[y,x] = (kernel*image_padded[y:y+3,x:x+3]).sum()   ## dot product of overlapping kernel and image
    return output
```

```python
from skimage import io, color

img = io.imread('../../Downloads/download.jpeg')    # Load the image
img = color.rgb2gray(img) 
```

```python
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]]) ## sharpen kernel as seen in the above image
image_sharpen = convolve2d(img,kernel)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(image_sharpen, cmap = plt.cm.gray)

## it's hard to sharpen shrek
```
